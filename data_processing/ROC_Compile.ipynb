```python
file_paths = [
    'LGBM ROC Curve with All Features.csv',
    'SVM ROC Curve with All Features.csv',
    'MLP ROC Curve with All Features.csv',
    'RF ROC Curve with All Features.csv',
    'KNN ROC Curve with All Features.csv',
]

file_paths_2 = [

    'LGBM ROC Curve with Important Features.csv',
    'SVM ROC Curve with Important Features.csv',
    'MLP ROC Curve with Important Features.csv',
    'RF ROC Curve with Important Features.csv',
    'KNN ROC Curve with Important Features.csv'


]
```

```python

def plot_roc_curves(file_path, title):
    plt.figure()

    for file_path in file_paths:
        # Generate label from file name without extension
        label = file_path.split('/')[-1].replace('.csv', '')

        # Load the CSV file
        df = pd.read_csv(file_path)

        # Extract true labels and predicted scores (adjust column names as needed)
        fpr = df['fpr']
        tpr = df['tpr']

        # Calculate ROC curve and AUC

        roc_auc = auc(fpr, tpr)

        # Plot ROC curve
        plt.plot(fpr, tpr, lw=2, label='%s (area = %0.2f)' % (label, roc_auc))

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()
```

```python
# all feature plot
plot_roc_curves(file_paths, "ROC Curve of All Models All Features")
```

```python
# selected feature plot
plot_roc_curves(file_paths_2, "ROC Curve of All Models Selected Features")
```
