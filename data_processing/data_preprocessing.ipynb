# Read and merge data

```python
valid = pd.read_csv("link_to_valid.csv")
valid
```

```python
fraud = pd.read_csv("link_to_fraud.csv")
fraud
```

```python
Features = ['domain', 'trust_score', 'blacklisted', 'ip_bad_count_total', 'ip_bad_count_bad', 'ssl_valid',
'ip_country', 'anonymous_whois', 'alexa_rank', 'dnsblock_threat_count', 'whois_registration_duration', 'whois_valid_email',
'avg_review_score', 'facebook_comments_total_count', 'facebook_comments_negative_count', 'facebook_comments_positive_count',
'popularity_rank','scamadviser_votes_legit', 'scamadviser_votes_scam', 'scamadviser_votes_fake', 'is_legit']

data = pd.merge(valid, fraud, on=Features, how='outer')
data
```

## Correcting missing data

```python
data['trust_score'].replace(-1, np.nan, inplace=True)
data['alexa_rank'].replace(-1, np.nan, inplace=True)
data['popularity_rank'].replace(-1, np.nan, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop(['alexa_rank', 'popularity_rank'], axis=1, inplace=True)
```

```python
data.info()
```

```python
print(data['ip_country'].mode()[0])
print(data['anonymous_whois'].mode()[0])
print(data['whois_valid_email'].mode()[0])
```

```python
data['trust_score'].fillna(data['trust_score'].mean(), inplace=True)
data['ip_country'].fillna(data['ip_country'].mode()[0], inplace=True)
data['anonymous_whois'].fillna(False, inplace=True)
data['whois_valid_email'].fillna(False, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['blacklisted'] = data['blacklisted'].astype(int)
data['ssl_valid'] = data['ssl_valid'].astype(int)
data['anonymous_whois'] = data['anonymous_whois'].astype(int)
data['whois_valid_email'] = data['whois_valid_email'].astype(int)
data['is_legit'] = data['is_legit'].astype(int)
```

```python
data.head()
```

```python
data.info()
```

### Create new columns

```python
# switch ip_bad_count_bad to a percentage of the total count
data['ip_bad_count_bad'] = np.where( data['ip_bad_count_total'] != 0, round(data['ip_bad_count_bad'] / data['ip_bad_count_total'], 2),
    np.where((data['ip_bad_count_total'] == 0) & (data['ip_bad_count_bad'] > 0), 1.00, 0.00)
)

#switch facebook comments to percentage of the total count
data['facebook_comments_negative_count'] = np.where(data['facebook_comments_total_count'] != 0, round(data['facebook_comments_negative_count'] / data['facebook_comments_total_count'], 2),
    np.where((data['facebook_comments_total_count'] == 0) & (data['facebook_comments_negative_count'] > 0), 1.00, 0.00)
)

data['facebook_comments_positive_count'] = np.where(data['facebook_comments_total_count'] != 0, round(data['facebook_comments_positive_count'] / data['facebook_comments_total_count'], 2),
    np.where((data['facebook_comments_total_count'] == 0) & (data['facebook_comments_positive_count'] > 0), 1.00, 0.00)
)
```

```python
data.drop(['ip_bad_count_total', 'facebook_comments_total_count'], axis=1, inplace=True)
```
